```python
import numpy as np
```

```python
def f(t, x):
#     return ((2*t)+x)*np.sin(2*t)
#     return -(1+x)/np.sin(t+1)
    return np.square(x) + t**(3/2)
```

```python
x = -1
t = 0
t_target = 2
h = 0.025
X = [x]
X_hat = [X[0]+h*f(t, X[0])][0:1]
```

```python
for i in range(int(t_target/h)-int(t/h)):
    X_hat.append(X[i] + h*f(t+i*h, X[i]))
    X.append(X[i] + 1/2*h*(f(t+i*h, X[i]) + f(t+(i+1)*h, X_hat[i+1])))

#     print(f'X_hat: {X_hat[i]}, X: {X[i]}')
```

```python
print(f'Therefore x({t_target}) ~ X({t_target}) = {X[int(t_target/h)]}')
```
